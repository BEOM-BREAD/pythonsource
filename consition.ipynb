```python
# 한 시간 임금은 10700원, 8시간 까지는 10700으로 지급하고 8시간 초과 근무시에는 1.5배지급

#금무시간(hours), rate(시간당임금), pay(일당)
hours, rate, pay = 12, 10700, 0

if hours > 8:
    pay = rate * 8 + ((hours - 8) * rate * 1.5)
else :
    pay = rate * hours

print("근무시간 : %d, 일당 : %d" % (hours, pay))
print("근무시간 : {0}, 일당 : {1:.0f}". format(hours, pay))
print(f"근무시간 : {hours}, 일당 : {pay:.0f}")
```

```python
# 1 ~ 100 사이의 숫자 중에서 3의 배수만을  더한 후 결과값 출력

i = 1
hap = 0
while i <= 100:
    if i % 3 == 0:
        hap = hap + i
    i = i + 1

print(f"3의 배수 합 {hap}")
```

```python
# 1 ~ 100 사이의 숫자 중에서 3의 배수와 5의 배수를 더한 후 결과값 출력

i = 1
hap = 0
while i <= 100:
    if i % 3 == 0 or i % 5 == 0:
        hap = hap + i
    i = i + 1

print(f"3의 배수, 5의 배수 합 {hap}")
```

```python
# 주차장 프로그램

# 사용자에게 메뉴 제공
# [1] 주차 | [2] 출차 | [3] 종료

# 사용자는 메뉴 선택
# 1
# 자동차 A ~ E
# A 자동차 입차. 주차장 상태 ==> ['A']
# 1
# B 자동차 입차. 주차장 상태 ==> ['A', 'B']
# 2
# B 자동차 출차, 주차장 상태 ==> ['A']
# E 자동차 출차, 주차장 상태 ==> ['A', 'B', 'C', 'D', 'E']
# 1
# 주차장 만차!

parking_lot = []
top, car_name = 0, 'A'

while True:
    no = int(input("[1] 주차 | [2] 출차 | [3] 종료"))

    if no <= 3:
        if no == 1:
            # top >= 5 주차장 만차
            if top >= 5:
                print("주차장 만차")
            else:
                parking_lot.append(car_name)
                # 출력 : A 자동차 입차, 주차장 상태 ==> ['A']
                print(f"{car_name} 자동차 입차. 주차장 상태 ==> {parking_lot}")
                car_name = chr(ord(car_name)+1)
                top = top + 1
        elif no == 2:
            if top > 0:
                # 리스트 뒤부터 제거 = pop()
                outCar = parking_lot.pop()
                print(f"{outCar} 자동차 출차. 주차장 상태 ==> {parking_lot}")
                top = top - 1
                car_name = chr(ord(car_name) - 1)
            else:
                print("빠져나갈 자동차가 없음")

            pass
        else:
            print("번호를 확인해 주세요")
            break
    else:
        print("번호를 확인해 주세요")

```
