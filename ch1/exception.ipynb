```python
# try ~ except ~ else

try:
    age = int(input("나이 입력"))
except:
    print("입력이 정확하지 않습니다.")
else: # 오류가 없다면
    if age <= 18:
        print("미성년자는 출입급지입니다.")
    else:
        print("환영합니다.")
```

```python
# 예외 클래스 생성 가능
class MyEerror(Exception):
    def __str__(self):
        return "허용되지 않는 별명입니다."


def say_nick(nick):
    if nick == "바보":
        raise MyEerror()
    print(nick)

try:  
    say_nick("천사")
    say_nick("바보")
except MyEerror as e:
    print(e)
```

```python
# 문제1. 랜덤 비밀번호 생성기

# 사용자로부터 생성할 비밀번호의 길이를 입력받아 랜덤 비밀번호를 생성하는 프로그램
# choice()
chars = "ABCDEFGHIJKLMNOPQRSTUVWXYZasbdefghijklnmopqrstuvwxyz1234567890!@#$%"
# 비밀번호의 길이는 4 이상 허용
# 숫자가 아닌 값을 입력하면 예외 처리
import random

while True:
    try:
        pwd_num = int(input("비밀먼호 길이 :"))

        if pwd_num < 4:
            print("4보단 커야지")
            continue
        break

    except ValueError:
        print("숫자를 입력해")

# 리스트 컴프리헨션
# password = []
# for i in range(pwd_num):
#     password.append(random.choice(chars))
password = [random.choice(chars) for _ in range(pwd_num)]

# print(f"비밀번호 :", password)
# 리스트를 => 문자열로 변경
print(f"비밀번호 : {"".join(password)}")
# try:
#     length = int(input("비밀번호 길이 입력(4글자 이상): "))

#     if length < 4:
#         print("비밀번호 길이는 4글자 이상이어야 합니다.")
#     else:
#         password = ""

#         for i in range(length):
#             password += random.choice(chars)

#         print("생성된 비밀번호 :", password)

# except ValueError:
#     print("숫자만 입력하세요.")
```

```python
# 나이 계산 프로그램
# 생년월일을 입력받아 현재 나이를 계산하는 프로그램 작성
# 입력 형식 2001-08-25 => 날짜 형식
import datetime

# 오늘날짜
now = datetime.datetime.now()

try:
    # 생년월일 입력
    birth_str = input("생년월일 입력 : (YYYY-MM-DD)")
    # '2001-05-15'
    birth = datetime.datetime.strptime(birth_str,"%Y-%m-%d")

    #2026 - 2021
    # now 년도만 추출
    age = now.year() - birth.year()
    if(now.month, now.day) < (birth.month,birth.day):
        age -= 1

    print("나이", age)
except:
    print("날짜 형식이 올바르지 않습니다")
```

```python
# 다음과 같은 csv 파일이 있다.
# 이름, 국어, 영어, 수학
# 홍길동, 90, 80, 70
# 김철수, 93, 83, 73
# 이영희, 94, 84, 74
# 성춘향, 95, 85, 75


import csv

with open("./data/student.csv", "w", encoding="utf-8", newline="") as f:
    headers = ['이름','국어','영어','수학']
    writer = csv.DictWriter(f, fieldnames=headers)

    # 헤더 작성
    writer.writeheader()
    writer.writerow({"이름" : "홍길동","국어" :90,"영어" :80,"수학" :70})
    writer.writerow({"이름" : "김철수","국어" :93,"영어" :83,"수학" :73})
    writer.writerow({"이름" : "이영희","국어" :94,"영어" :84,"수학" :74})
    writer.writerow({"이름" : "성춘향","국어" :95,"영어" :85,"수학" :75})

# csv 파일을 읽어 학생별 평균 점수 출력
# 홍길동 : 80.0

# csv.DictReader() 사용, 평균은 둘째자리까지

with open("./data/student.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    # print(reader)

    for row in reader:
        # print(row)  # {'이름': '홍길동', '국어': '90', '영어': '80', '수학': '70'}
        avg = (int(row.get('국어')) + int(row.get('영어')) +int(row.get('수학'))) / 3
        print(f"{row.get('이름')} : {avg:.2f}")
```
