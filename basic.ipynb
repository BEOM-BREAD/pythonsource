```python
# http://naver.com

# 비밀번호 생성
# 규칙 1. https:// 제외
# 규칙 2. 처음 만나는 . 이후 부분은 제외 => naver
# 규칙 3. 남은 글자 중 처음 세자리 + 글자 갯수 + 글자 내 'e' 개수 + '!' 비밀번호 생성 => nav51!

# 문자 함수, 슬라이싱, str()

url = "http://naver.com"

# password = url[7:13]
password = url.replace("http://","")

password = password[:password.find(".")]
password = password[:3] + str(len(password)) + str(password.count('e')) + '!'
password
```
